```python
import requests
import random
from bs4 import BeautifulSoup as bs
from selenium import webdriver
```

```python
#headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36 Edg/91.0.864.59'}
#page = requests.get("https://ca.finance.yahoo.com/quote/OTEX?p=OTEX",headers=headers)
driver=webdriver.Edge('./msedgedriver.exe')
url="https://ca.finance.yahoo.com/quote/OTEX?p=OTEX"
driver.get(url)
soup = bs(driver.page_source, 'html.parser')
print(soup.prettify())
```

```python
quote_summary_table=soup.find(id="quoteNewsStream-0-Stream")
print(quote_summary_table.prettify())
```

```python
table_rows = quote_summary_table.find_all('a')
print(table_rows)
#ad_name= quote_summary_table.find_all(class_=)
```

```python
text = []
value=[]
for row in (table_rows):
    text.append(row.text)
    value.append(row['href'])
```
